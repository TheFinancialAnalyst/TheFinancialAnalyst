```python
# Filter the relevant columns
ratios = df[['PBITM (%)', 'PBDTM (%)', 'CPM (%)', 'APATM (%)', 'ROCE (%)', 'RONW (%)', 'Payout (%)']]

# Create subplots
fig, axes = plt.subplots(2, 4, figsize=(15, 8))

# Plot pie charts for each ratio
for i, (ratio, ax) in enumerate(zip(ratios.columns, axes.flatten())):
    data = df.groupby('Year').last()[ratio]
    ax.pie(data, labels=data.index, autopct='%1.1f%%')
    ax.set_title(ratio)

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
