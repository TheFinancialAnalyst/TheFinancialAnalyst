We are predicting the values using statsmodel and sciket.learn and moreover we are not taking the individual values and after that we have found the values.

```python
import statsmodels.api as sm
x = sm.add_constant(data[['Year', 'Revenue From Operations', 'Less: Excise Duty',
       'Revenue From Operations(Net)', 'Other Income', 'Total Revenue',
       'Cost of Material Consumed', 'Purchases of Stock-in-Trade',
       'Changes in Inventories of Finished Goods, Work-in-Progress and Stock-in-Trade',
       'Employee Benefits / Salaries & other Staff Cost', 'Finance Cost',
       'Depreciation and Amortization', 'Other Expenses', 'Total Expenses',
       'Profit Before Exceptional Items and Tax',
       'Profit Before Extraordinary Items and Tax', 'Profit Before Tax',
       'Tax Expenses', 'Current Tax', 'Deferred Tax',
       'Profit After Pre-acquisition Profit',
       'Profit/(Loss) for the period from Continuing Operations',
       'Profit Attributable to Shareholders',
       'Profit Attributable to Ordinary Shareholders',
       'Profits Available for Appropriation',
       'Unappropriated Profits Carried Forward',
       'Other Comprehensive Income That Will Be Reclassified to Profit Or Loss :',
       'Total Comprehensive Income for the Year', 'Earning Per Share - Basic',
       'Earning Per Share - Diluted', 'Final', 'Final.1',
       'Total Dividend Percentage', 'Total Dividend',
       'Weighted Average Number of Shares in Issue - Basic',
       'Weighted Average Number of Shares in Issue - Diluted',
       'Final Dividend Amount']])
x.head()
```

```python
y = data['Profit After Tax']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=100)
x_train
```

```python
train_test_split
```

```python
from IPython.display import display
```

```python
data_lm = sm.OLS(y_train, x_train).fit()
display( data_lm.params )
```

```python
data_lm.summary()
```

```python
%matplotlib inline
# Generate some random data
data_resid = np.random.normal(size=100)

# Create the ProbPlot object and the P-P plot
probplot = sm.ProbPlot(data_resid)
fig = probplot.ppplot(line='45')

# Set the title of the plot
plt.title("Normal P-P Plot of Mahindra & Mahindra Regression Standardized")

# Show the plot
plt.show()
```

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
resid = data_resid[data_lm.fittedvalues.index]
# Plot the scatter plot
plt.scatter(get_standardized_values(data_lm.fittedvalues), get_standardized_values(resid))
plt.title("Residual Plot: Mahindra & Mahindra P&L Prediction")
plt.xlabel("Standardized predicted values")
plt.ylabel("Standardized residuals")
plt.show()
```

```python
from scipy.stats import zscore
```

```python
data_influence = data_lm.get_influence()
(c, p) = data_influence.cooks_distance

plt.stem(np.arange( len( x_train) ),
         np.round( c, 3),
         markerfmt=",");

plt.title( "Cooks distance for all observations in M&M P&L data set" );

plt.xlabel("Row index")

plt.ylabel("Cooks Distance");
```

```python
from statsmodels.graphics.regressionplots import influence_plot

fig, ax = plt.subplots( figsize=(8,6))
influence_plot( data_lm, ax = ax)
plt.title("M&M P&L Leverage Value Vs Residuals")
plt.show();
```
